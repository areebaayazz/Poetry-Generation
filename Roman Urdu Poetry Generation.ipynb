```python
import numpy as np
import spacy
import random
import csv
import requests
import re
```

```python
#Reading the poetry text file
with open('poetry.txt', 'rt', encoding="utf-8") as file:
    csv_reader = csv.reader(file)
    poetry = list(csv_reader)

#print(poetry)
#printing only 11 lines
poetry[:11]
```

```python
#tokenize poetry corpus into individual words

#loading the urdu model
Ud = spacy.blank('ur')

#array to store
words = []

# #poetry has a list of strings, so we join it
# P2 = '\n'.join(poetry)

# doc = Ud(P2)   
# for i in doc:
#     words.append(i.text)
# words

for i in poetry:
    #if i is a string, then add to words array
    if isinstance(i, str):
        doc = Ud(i)
        for j in doc:
            words.append(j.text)
    #if not a string, then join and tokenize
    elif isinstance(i, list):
        sent = ' '.join(i)
        doc = Ud(sent)
        for j in doc:
            words.append(j.text)
words
```

```python
#scraping this website to add more words into the corpus
url = 'https://hamariweb.com/poetry/urdu-ghazals/'

response = requests.get(url)
html = response.text

#regular expression to store all urdu words
urdu_words = re.findall(r'[\u0600-\u06FF]+', html)

urdu_words

len(urdu_words)
```

```python
#joining the poetry corpus with the corpus made from web scraping
words.extend(urdu_words)
words
```

```python
#to store the starting and ending words of stanzas in poetry

corpus = [] #to store all lines in file poetry
start = []
end = []

for stanza in poetry: #loop all stanzas(before blank line)
    for line in stanza: #loop each line in stanza
        #to check if line is not empty
        if(line.strip()): 
            corpus.append(line.strip())
            #split line into words
            word_list = line.split()
            if(word_list):
                #get starting and ending words
                start.append(word_list[0])
                end.append(word_list[-1])
start
```

### Unigram Model

```python
unigrams = []
count = 0
total = len(words) #25814

for w in words:
    if w != -1:
        count = words.count(w)
        probability = count / total
        unigrams.append((w, probability))

print("The length of unigrams is:", len(unigrams))
print("Example:", unigrams[202]) #prob of one example word
```

### Bigram Model

```python
total
```

```python
bigrams = []

for i in range(total - 1):
    curr = words[i]
    nex = words[i+1]
    c1 = words.count(curr)
    c2 = words.count(nex)
    count = 0
    
    for j in range(total - 1):
        if (words[j] == curr) and (words[j+1] == nex):
            count += 1
            
    # find prob of bigram
    prob = count / c1
    
    # add the bigram words and their respective prob to list
    bigrams.append((curr, nex, prob))
```

### Trigram Model

```python
trigrams = []

for i in range(total - 2):
    cur = words[i]
    n1 = words[i+1]
    n2 = words[i+2] #next-next word

    if (cur != -1 and n1 != -1 and n2 != -1): #check to see if any words are -1
        count = 0 #count the freq of trigram
        prev = 0
        for j in range(total - 2):
            if (words[j] == cur and words[j+1] == n1 and words[j+2] == n2):
                count += 1
            if (words[j] == n1 and words[j+1] == n2):
                prev += 1
                
        #prob of trigram w.r.t count and prev
        prob = 0 if prev == 0 else count / prev
        
        trigrams.append((n2, cur, n1, prob))
```

```python
print("Example:", trigrams[113])
```

### Backward Bigram

```python
backwardbi = []

# Loop through all words except the last word
for i in range(total - 1):
    count = 0
    curr = words[i]
    nex = words[i+1]

    #check if current and next word are not -1
    if(curr != -1 and nex != -1):
        loc = total - 1  
        while loc > i+1:    # limit the loop to search backward from word[i+1] to the beginning of word
            # If the previous and current words match, increment count
            if words[loc] == nex and words[loc - 1] == curr:
                count += 1
            loc -= 1

        #find prob only if nex is present in the list
        if words.count(nex) > 0:
            prob = count / words.count(nex)
        else:
            prob = 0

        backwardbi.append((curr,nex,prob))
```

# Documentation:

#### Read a text file called "poetry.txt" containing the poetry corpus and open it
#### An empty list "words" is initialized to store individual words.
#### A for loop iterates over each element in the "poetry" list:
#### If the element is a string, it is tokenized using the loaded Urdu model and appended to "words".
#### If the element is a list, it is joined into a sentence and tokenized in the same way.
#### The "urdu_words" list is appended to the "words" list after web scraping
#### Three empty lists are created: "corpus", "start", and "end".
#### A for loop iterates over each stanza in the "poetry" corpus:
#### A nested for loop iterates over each line in the current stanza.
#### If the line is not empty, it is appended to "corpus" and the first and last words are appended to "start" and "end", respectively.
#### An empty list "unigrams" is initialized to store unique words and their probabilities.
#### A counter "count" is set to 0 and the length of "words" is stored in the variable "total".
#### A for loop iterates over each word in "words":
#### If the word is not equal to -1, the count of its occurrences is calculated and its probability is appended as a tuple to the "unigrams" list.
#### An empty list "bigrams" is initialized to store unique bigrams and their probabilities.
#### A for loop iterates over each word in "words", except for the last word:
#### The current word and the next word are stored as "curr" and "nex", respectively.
#### The frequency of "curr" and "nex" in the "words" list are counted and stored as "c1" and "c2".
#### A counter "count" is initialized to 0.
#### Another for loop iterates over each pair of adjacent words in the "words" list. If the current pair of words match "curr" and "nex", then "count" is incremented.
#### The probability of the bigram "curr nex" is calculated and appended as a tuple to the "bigrams" list
#### these models are used to generate poetry
